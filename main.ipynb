```python
import pandas as pd
import numpy as np
```

```python


movies = pd.read_csv('./dataset/movies_metadata.csv', low_memory = False)
movies.head(3)
```

```python
ratings = pd.read_csv('./dataset/ratings.csv')
ratings.head()
```

```python
movies = movies[['id', 'original_title', 'original_language', 'vote_count']]

movies = movies.rename (
    columns = {
        'id': 'MOVIE_ID',
        'original_title': 'TITLE',
        'original_language': 'LANGUAGE',
        'vote_count': 'VOTE_COUNT'
    }
)

movies.head()
```

```python
ratings = ratings[['userId', 'movieId', 'rating']]

ratings = ratings.rename (
    columns = {
        'userId': 'USER_ID',
        'movieId': 'MOVIE_ID',
        'rating': 'RATING'
    }
)

ratings.head()
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace = True)
```

```python
ratings.isna().sum()
```

```python
ratings['USER_ID'].value_counts()
```

```python
rating_count = ratings['USER_ID'].value_counts() > 999
y = rating_count[rating_count].index
y.shape
```

```python
ratings.shape
```

```python
ratings = ratings[ratings['USER_ID'].isin(y)]
ratings.shape
```

```python
ratings.head()
```

```python
movies = movies[movies['VOTE_COUNT'] > 999]
```

```python
language_movies = movies['LANGUAGE'].value_counts()
language_movies.head(20)
```

```python
movies = movies[movies['LANGUAGE'] == 'en']
```

```python
movies.info()
```

```python
ratings.info()
```

```python
movies['MOVIE_ID'] = movies['MOVIE_ID'].astype(int)
```

```python
movies.shape
```

```python
df = ratings.merge(movies, on = 'MOVIE_ID')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.drop_duplicates(['USER_ID', 'MOVIE_ID'], inplace = True)
df.shape
```

```python
del movies['MOVIE_ID']
```

```python
movies_pivot = df.pivot_table(columns = 'USER_ID', index = 'TITLE', values = 'RATING')

movies_pivot.head()
```

```python
movies_pivot.fillna(0, inplace = True)
movies_pivot.head()
```

```python
from scipy.sparse import csr_matrix

movies_sparse = csr_matrix(movies_pivot)
```

```python
type(movies_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(algorithm = 'brute')
model.fit(movies_sparse)
```

```python
def getSuggestions(movie_name):
    _, suggestions_id = model.kneighbors(movies_pivot.filter(items = [movie_name], axis = 0).values.reshape(1, -1))
    movie_list = [movie for movie in movies_pivot.index[suggestions_id[0]]]
    return movie_list[1:]
```

```python
from random import choice
```

```python
choice(getSuggestions('Toy Story'))
```

```python
choice(getSuggestions('1408'))
```

```python
choice(getSuggestions('2 Fast 2 Furious'))
```
